## 为什么训练误差比测试误差高很多？

Keras模型有两种模式：训练和测试。正则化机制，如Dropout和L1/L2权重正则化，在测试是关闭的。  
此外，训练误差是每批训练数据的平均误差。由于你的模型是随着时间而变化的，一个epoch中的第一批数据的误差通常比最后一批的要高。另一方面，测试误差是模型在一个epoch训练完成后计算的，因而误差较小。
