```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import Dropout
```

```python
# define parameters
num_genes = 1000
num_cell_types = 25
input_size = (num_genes,)
epochs = 50
batch_size = 32
```

```python
model = Sequential()
```

```python
# constract model

# firsr layer
model.add(Dense(100,
                input_shape=input_size,
                # weight addition regularization, just like original author
                kernel_regularizer=keras.regularizers.l2(0.005),
                # initialize weight with Xavier
                kernel_initializer=keras.initializers.glorot_normal(seed=3),
                # initialize bias with Xavier
                bias_initializer=keras.initializers.Zeros()))
model.add(Activation('relu'))

# second layer
model.add(Dense(50,
                kernel_regularizer=keras.regularizers.l2(0.005),
                kernel_initializer=keras.initializers.glorot_normal(seed=3),
                bias_initializer=keras.initializers.Zeros()))
model.add(Activation('relu'))

# third layer
model.add(Dense(num_cell_types,
                kernel_regularizer=keras.regularizers.l2(0.005),
                kernel_initializer=keras.initializers.glorot_normal(seed=3),
                bias_initializer=keras.initializers.Zeros()))
model.add(Activation('softmax'))
```

```python
adam = keras.optimizers.Adam(lr=0.0001, decay=0.95)
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer=adam,
             metrics=['accuracy'])
```
